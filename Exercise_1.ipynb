```python
Edited by Joydeep's another account
# Import necessary libraries like
import pandas as pd
```

```python
# Read the Excel file into a pandas DataFrame
# Replace 'path_to_file.xlsx' with the actual path to your file
file_path = '/content/drive/MyDrive/DO Lab/question_1.xlsx'
df = pd.read_excel(file_path, engine='openpyxl')
```

```python
# Display the top 5 entries(rows) from the dataset
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
# Create new columns for Total, Maximum, Minimum, and Average marks
df['Total'] = df[['math score', 'reading score', 'writing score']].sum(axis=1)
df['Maximum'] = df[['math score', 'reading score', 'writing score']].max(axis=1)
df['Minimum'] = df[['math score', 'reading score', 'writing score']].min(axis=1)
df['Average'] = df[['math score', 'reading score', 'writing score']].mean(axis=1)
```

```python

import matplotlib.pyplot as plt

# Assuming df is your DataFrame and it already contains the 'Total' column
# Create the histogram
df['Total'].plot(kind='hist', bins=10, color='skyblue', edgecolor='black', title='Histogram of Total Marks')

# Add labels
plt.xlabel('Total Marks')
plt.ylabel('Frequency')

# Show the plot
plt.show()
```
